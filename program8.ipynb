Crisp partitions for real-life iris dataset

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Load Iris Dataset
iris = load_iris()
X = iris.data
y_true = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Evaluate Clusters vs True Labels
print("Crisp Clustering Confusion Matrix:\n")
print(confusion_matrix(y_true, y_kmeans))

# Visualize Clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X[:, 2], y=X[:, 3], hue=y_kmeans, palette='Set1', style=y_true, s=100)
plt.title("K-Means Crisp Partitioning of Iris Data")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```
